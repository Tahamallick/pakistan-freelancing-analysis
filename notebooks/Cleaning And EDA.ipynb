```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv('pk_freelancing_economy_2015_2025.csv')
```

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nFirst look:")
df.head(10)
```

```python
df['review_score']=df['review_score'].fillna(df['review_score'].median())
```

```python
df['job_success_score_pct']=df['job_success_score_pct'].fillna(df['job_success_score_pct'].median())
```

```python
df.isnull().sum()
```

```python
len(df)
```

```python
repeatedclient=df.groupby('skill_category')['repeat_client_rate_pct'].mean().sort_values(ascending=False)
```

```python
repeatedclient
```

```python
skillspaywellinusd=df.groupby('skill_category')['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
skillspaywellinusd
```

```python
cityearningwellinusd=df.groupby('city')['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
cityearningwellinusd
```

```python
mostearning_platform_in_usd=df.groupby('platform')['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
mostearning_platform_in_usd
```

```python
gender_earning_most_in_usd=df.groupby('gender')['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
gender_earning_most_in_usd
```

```python
profile_level_wise_client_repeat=df.groupby('profile_level')['repeat_client_rate_pct'].mean().sort_values(ascending=False)
```

```python
profile_level_wise_client_repeat
```

```python
education_level_wise_earning=df.groupby('education_level')['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
education_level_wise_earning
```

```python
df['exp_bracket'] = pd.cut(df['experience_years'], 
                            bins=[0, 1, 3, 5, 8, 20], 
                            labels=['0-1 yr', '1-3 yrs', '3-5 yrs', '5-8 yrs', '8+ yrs'])
exp_earning = df.groupby('exp_bracket', observed=True)['monthly_earnings_usd'].mean().sort_values(ascending=False)
```

```python
exp_earning
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Pakistan Freelancing Economy 2015-2025', fontsize=20, fontweight='bold', y=1.02)

axes[0,0].barh(skillspaywellinusd.index, skillspaywellinusd.values, color='steelblue')
axes[0,0].set_title('Average Monthly Earnings by Skill (USD)', fontweight='bold')
axes[0,0].set_xlabel('Avg Monthly Earnings (USD)')

yearly = df.groupby('year')['monthly_earnings_usd'].mean()
axes[0,1].plot(yearly.index, yearly.values, marker='o', color='green', linewidth=2)
axes[0,1].set_title('Yearly Earnings Trend (USD)', fontweight='bold')
axes[0,1].set_xlabel('Year')
axes[0,1].set_ylabel('Avg Earnings (USD)')

axes[1,0].bar(mostearning_platform_in_usd.index, mostearning_platform_in_usd.values, color=['#1f77b4','#ff7f0e','#2ca02c'])
axes[1,0].set_title('Average Earnings by Platform (USD)', fontweight='bold')
axes[1,0].set_ylabel('Avg Monthly Earnings (USD)')

top_levels = profile_level_wise_client_repeat.head(7)
axes[1,1].barh(top_levels.index, top_levels.values, color='coral')
axes[1,1].set_title('Repeat Client Rate by Profile Level (%)', fontweight='bold')
axes[1,1].set_xlabel('Repeat Client Rate (%)')

axes[2,0].bar(gender_earning_most_in_usd.index, gender_earning_most_in_usd.values, color=['#e377c2','#1f77b4'])
axes[2,0].set_title('Average Earnings by Gender (USD)', fontweight='bold')
axes[2,0].set_ylabel('Avg Monthly Earnings (USD)')
axes[2,0].set_ylim(1600, 1800)

axes[2,1].barh(repeatedclient.index, repeatedclient.values, color='mediumpurple')
axes[2,1].set_title('Repeat Client Rate by Skill (%)', fontweight='bold')
axes[2,1].set_xlabel('Repeat Client Rate (%)')

plt.tight_layout()
plt.savefig('pakistan_freelancing_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
df.to_csv('freelancing_cleaned.csv', index=False)

skills_summary = df.groupby('skill_category').agg(
    avg_earnings_usd=('monthly_earnings_usd', 'mean'),
    avg_hourly_rate=('hourly_rate_usd', 'mean'),
    avg_repeat_client=('repeat_client_rate_pct', 'mean'),
    total_freelancers=('freelancer_id', 'count')
).reset_index()

city_summary = df.groupby('city').agg(
    avg_earnings_usd=('monthly_earnings_usd', 'mean'),
    avg_repeat_client=('repeat_client_rate_pct', 'mean'),
    total_freelancers=('freelancer_id', 'count')
).reset_index()

yearly_summary = df.groupby('year').agg(
    avg_earnings_usd=('monthly_earnings_usd', 'mean'),
    avg_pkr_usd_rate=('pkr_usd_rate', 'mean'),
    avg_earnings_pkr=('monthly_earnings_pkr', 'mean'),
    total_freelancers=('freelancer_id', 'count')
).reset_index()

platform_summary = df.groupby('platform').agg(
    avg_earnings_usd=('monthly_earnings_usd', 'mean'),
    avg_repeat_client=('repeat_client_rate_pct', 'mean'),
    total_freelancers=('freelancer_id', 'count')
).reset_index()

skills_summary.to_csv('summary_skills.csv', index=False)
city_summary.to_csv('summary_city.csv', index=False)
yearly_summary.to_csv('summary_yearly.csv', index=False)
platform_summary.to_csv('summary_platform.csv', index=False)

```
